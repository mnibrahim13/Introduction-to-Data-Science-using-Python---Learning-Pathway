# Data Visualization with Python: Plotting Dots and Lines Using Matplotlib

### Introduction to Data Visualization

Data visualization is a critical component of data analysis. It allows you to understand data patterns, trends, and relationships by representing data in a visual context. This is particularly helpful in identifying correlations, outliers, and distributions that are not immediately apparent from raw data alone.

Matplotlib is one of the most popular Python libraries for creating static, animated, and interactive visualizations in Python. It provides a wide variety of chart types to help communicate data effectively.

```python
# Importing Matplotlib
# Import the `pyplot` module from matplotlib, which provides a convenient interface for plotting. It is conventionally imported as `plt`.
import matplotlib.pyplot as plt
import numpy as np
```

### Creating Basic Line Plots

A line plot is used to represent continuous data and is often used to show changes over time.

```python
# Creating Data
x = np.linspace(0, 10, 100)  # 100 evenly spaced points between 0 and 10
y = np.sin(x)  # Sine of each point in x
# Plotting the Line Graph
plt.figure(figsize=(10, 5))  # Setting figure size
plt.plot(x, y)  # Creating the line plot
plt.title("Basic Line Plot")  # Adding a title
plt.xlabel("X-axis")  # Adding X-axis label
plt.ylabel("Y-axis")  # Adding Y-axis label
plt.grid(True)  # Adding grid lines
plt.show()  # Displaying the plot
```

We can add customizations like titles, labels, legends, grid lines, and axis limits to make our plots more informative and aesthetically pleasing.

```python
# Creating and Customizing a Line Plot
plt.figure(figsize=(10, 5))
y2 = np.cos(x)
plt.plot(x, y, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='red', linestyle='--')  # Dashed line
plt.title("Sine and Cosine Waves")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Adding a legend
plt.grid(True)
plt.show()
```

### Scatter Plots (Plotting Dots)

A scatter plot is useful when we need to visualize the relationship between two variables. Unlike line plots, scatter plots do not connect the points with lines.

```python
# Creating Data for Scatter Plot
x = np.random.rand(50)  # 50 random points between 0 and 1
y = np.random.rand(50)

# Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='green', marker='o')
plt.title("Scatter Plot of Random Points")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

### Plot Customization (Colors, Styles, Markers)

You can use various arguments to customize the plots. Some key customizations include:
- **Colors**: You can set the color of lines and markers (e.g., 'red', '#00ff00').
- **Markers**: You can use markers like 'o', '^', '*', etc., to highlight points.
- **Line Styles**: You can use line styles such as solid ('-'), dashed ('--'), or dotted (':').

### Combining Line and Scatter Plots

```python
# Combining Line and Scatter Plots
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Line (Sine)', color='blue', linestyle='-')
plt.scatter(x, y, color='orange', marker='o', label='Scatter Points')
plt.title("Combined Line and Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Map target numbers to species names
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
iris_df['species']
```

```python
# Plotting the distribution of sepal length for each species
plt.figure(figsize=(10, 6))
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    plt.hist(subset['sepal length (cm)'], alpha=0.5, label=species)

plt.title('Distribution of Sepal Length for Each Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Scatter plot of sepal length vs sepal width
plt.figure(figsize=(10, 6))
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species)

plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

Box plot

- The box represents the interquartile range (IQR), which contains the middle 50% of the data points.
- The horizontal line inside the box represents the median (the middle value).
- The whiskers (lines extending from the box) represent the minimum and maximum values within a specified range (Generally 1.5 times IQR).
- Circles outside the whiskers represent outliers—data points that fall significantly outside the general distribution.

```python
# Box plot for petal length for each species
plt.figure(figsize=(10, 6))
iris_df.boxplot(column='petal length (cm)', by='species', grid=False)
plt.title('Box Plot of Petal Length by Species')
plt.suptitle('')  # Removes the automatic 'Boxplot grouped by...' title
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Box plot for sepal length for each species
plt.figure(figsize=(10, 6))
iris_df.boxplot(column='sepal length (cm)', by='species', grid=False)
plt.title('Box Plot of Petal Length by Species')
plt.suptitle('')  # Removes the automatic 'Boxplot grouped by...' title
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
iris_df.groupby('species')['petal length (cm)'].head()
```

```python
# Line plot to visualize average sepal length of each species
avg_sepal_length = iris_df.groupby('species')['sepal length (cm)'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_sepal_length.index, avg_sepal_length.values, marker='o', linestyle='-', color='b')
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.grid(True)
plt.show()
```
